# Nova seção

```python
import pandas as pd
import numpy as np

np.random.seed(0)
n_rows = 1000

data = {
    'Data': pd.date_range(start='2023-01-01', periods=n_rows, freq='D'),
    'Produto': np.random.choice(['monitor', 'notebook', 'teclado', 'mouse', 'headset', 'placa de Video', 'Tablet', 'SmartFone'], n_rows),
    'Categoria': np.random.choice(['Eletronico', 'Acessórios'], n_rows),
    'Quantidade': np.random.randint(1, 10, n_rows),
    'Preço': np.random.uniform(100, 3000, n_rows),
    'Valor': np.random.randint(100, 1000, n_rows),
    'Client_Id': np.random.randint(1000, 9999, n_rows),
    'Cidade': np.random.choice(['São Paulo', 'Rio de Janeiro', 'Belo Horizonte', 'Porto Alegre', 'Curitiba'], n_rows),
    'Estado': np.random.choice(['SP', 'RJ', 'MG', 'RS', 'PR'], n_rows)
}

df = pd.DataFrame(data)

df['Valor_total'] = df['Preço'] * df['Quantidade']
print("Cabeçalho")
print(df.head())
print("")
print("")

print("Info basica do Data Frame")
print(df.info())
print("")
print("")
print("Estatistica descritiva")
print(df.describe())
print("")
print("")

vendas_por_produtos = df.groupby('Produto')['Valor_total'].sum().sort_values(ascending=False)
print("Vendas por Produtos")
print(vendas_por_produtos)
print("")
print("")
 # Ordena os valores do maior para o menor (por padrão, ascending=True ordenaria do menor para o maior).

vendas_por_cidade = df.groupby('Cidade')['Valor_total'].sum().sort_values(ascending=False)
print("Vendas por Cidade")
print(vendas_por_cidade)
print("")
print("")

vendas_maior_valor = df[df['Valor_total'] > 1000]
print("Vendas com valor maior que 1000")
print(vendas_maior_valor)
print("")
print("")
 #set_index = Define a coluna 'Data' como índice do DataFrame.  resample = Reamostra os dados em uma nova frequência de tempo.
vendas_mensais = df.set_index('Data').resample('M')['Valor_total'].sum()
print("Vendas Mensais")
print(vendas_mensais)
print("")
print("")
# df.drop_duplicates(inplace=True)
df = df.round({'Preço': 2, 'Valor_total': 2})
# df.to_csv('testepandas.csv', index=False)


print(df)
```

```python
import matplotlib.pyplot as plt
import numpy
notas = np.random.normal(70, 15,1000)

plt.figure(figsize=(10, 6 ))
plt.hist(notas, bins=20, edgecolor='black')
plt.title('Distribuição de Notas')
plt.xlabel('Notas')
plt.ylabel('Frequência')
plt.show()
```
